```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
wt_energy = np.loadtxt('7BZ5_WT/production_new/namdenergy/Energy_output_total')
uk_energy = np.loadtxt('7BZ5_N501Y/production_new/namdenergy/Energy_output_total')
saf_energy = np.loadtxt('7BZ5_South_African/production-new/namdenergy/Energy_output_total')
ind_energy = np.loadtxt('7BZ5_Indian/production_new/namdenergy/Energy_output_total')
delta_energy = np.loadtxt('7BZ5_delta/production/namdenergy/Energy_output_total')
```

```python
nsm = 10 #number of frames to smooth 
wt_energy_smooth = np.zeros(len(wt_energy))
wt_energy_err = np.zeros(len(wt_energy))
for i in range(len(wt_energy)):
    wt_energy_smooth[i] = np.mean(wt_energy[i-nsm:i+nsm,4])
    wt_energy_err[i] = np.std(wt_energy[i-nsm:i+nsm,4])
    
wt_electro_smooth = np.zeros(len(wt_energy))
wt_electro_err = np.zeros(len(wt_energy))
for i in range(len(wt_energy)):
    wt_electro_smooth[i] = np.mean(wt_energy[i-nsm:i+nsm,2])
    wt_electro_err[i] = np.std(wt_energy[i-nsm:i+nsm,2])
    
wt_vdw_smooth = np.zeros(len(wt_energy))
wt_vdw_err = np.zeros(len(wt_energy))
for i in range(len(wt_energy)):
    wt_vdw_smooth[i] = np.mean(wt_energy[i-nsm:i+nsm,3])
    wt_vdw_err[i] = np.std(wt_energy[i-nsm:i+nsm,3])
```

```python
nsm = 10 #number of frames to smooth 
delta_energy_smooth = np.zeros(len(delta_energy))
delta_energy_err = np.zeros(len(delta_energy))
for i in range(len(delta_energy)):
    delta_energy_smooth[i] = np.mean(delta_energy[i-nsm:i+nsm,4])
    delta_energy_err[i] = np.std(delta_energy[i-nsm:i+nsm,4])
    
delta_electro_smooth = np.zeros(len(delta_energy))
delta_electro_err = np.zeros(len(delta_energy))
for i in range(len(delta_energy)):
    delta_electro_smooth[i] = np.mean(delta_energy[i-nsm:i+nsm,2])
    delta_electro_err[i] = np.std(delta_energy[i-nsm:i+nsm,2])
    
delta_vdw_smooth = np.zeros(len(delta_energy))
delta_vdw_err = np.zeros(len(delta_energy))
for i in range(len(delta_energy)):
    delta_vdw_smooth[i] = np.mean(delta_energy[i-nsm:i+nsm,3])
    delta_vdw_err[i] = np.std(delta_energy[i-nsm:i+nsm,3])
```

```python
x = wt_energy[:,0]+101
plt.plot(x,wt_energy_smooth,'g',label='WT')
plt.fill_between(x, wt_energy_smooth - wt_energy_err, wt_energy_smooth + wt_energy_err, facecolor='xkcd:lightgreen')

x = uk_energy[:,0]+101
plt.plot(x,uk_energy_smooth,'b',label='UK Variant')
plt.fill_between(x, uk_energy_smooth - uk_energy_err, uk_energy_smooth + uk_energy_err, facecolor='xkcd:lightblue')

x = saf_energy[:,0]+101
plt.plot(x,saf_energy_smooth,'r',label='SA Variant')
plt.fill_between(x, saf_energy_smooth - saf_energy_err, saf_energy_smooth + saf_energy_err, facecolor='pink')

x = ind_energy[:,0]+101
plt.plot(x,ind_energy_smooth,'k',label='Ind Variant')
plt.fill_between(x, ind_energy_smooth - ind_energy_err, ind_energy_smooth + ind_energy_err, facecolor='gray',alpha=0.5)


x = delta_energy[100:,0]
plt.plot(x,delta_energy_smooth[100:],'orange',label='delta Variant')
plt.fill_between(x, delta_energy_smooth[100:] - delta_energy_err[100:], delta_energy_smooth[100:] + delta_energy_err[100:], facecolor='orange',alpha=0.5)


plt.legend(fontsize=14,ncol=2)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.ylim(-450)
plt.xlabel('Time (ns)',fontsize=16)
plt.ylabel('Non-bonded Energy (kcal/mol)',fontsize=16)

plt.savefig('energy-figures/7BZ5_Nonbonded_energy.png',dpi=200,bbox_inches='tight')
```
