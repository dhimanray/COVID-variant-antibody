```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
wt = np.loadtxt('7BZ5_WT/production_new/MM-GBSA/freeE_each_frame.dat')
uk = np.loadtxt('7BZ5_N501Y/production_new/MM-GBSA/freeE_each_frame.dat')
saf = np.loadtxt('7BZ5_South_African/production-new/MM-GBSA/freeE_each_frame.dat')
ind = np.loadtxt('7BZ5_Indian/production_new/MM-GBSA/freeE_each_frame.dat')
delta = np.loadtxt('7BZ5_delta/production/MM-GBSA/freeE_each_frame.dat')
```

```python
labels = ['WT','N501Y','K417N\nE484K\nN501Y','L452R\nE484Q','L452R\nT478K']
mmgbsa_pt = [np.mean(wt),np.mean(uk),np.mean(saf),np.mean(ind),np.mean(delta)]
mmgbsa_err = [np.std(wt),np.std(uk),np.std(saf),np.std(ind),np.std(delta)]
colors = ['g','b','r','gray','orange']

plt.bar(labels,mmgbsa_pt,yerr=mmgbsa_err,capsize=20.0,color=colors)
plt.xticks(rotation=60)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.title('B38 + RBD',fontsize=16)
plt.ylabel('$\Delta G_b$ (kcal/mol)',fontsize=16)
plt.savefig('mmgbsa-plot/free-energy.pdf',bbox_inches='tight')
```
