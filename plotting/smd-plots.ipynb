```python
wt_all = np.array([wt1[-1,4],wt2[-1,4],wt3[-1,4],wt4[-1,4],wt5[-1,4]])

wt_mean = -np.log(np.mean(np.exp(-wt_all)))


uk_all = np.array([uk1[-1,4],uk2[-1,4],uk3[-1,4],uk4[-1,4],uk5[-1,4]])

uk_mean = -np.log(np.mean(np.exp(-uk_all)))

saf_all = np.array([saf1[-1,4],saf2[-1,4],saf3[-1,4],saf4[-1,4],saf5[-1,4]])

saf_mean = -np.log(np.mean(np.exp(-saf_all)))

ind_all = np.array([ind1[-1,4],ind2[-1,4],ind3[-1,4],ind4[-1,4],ind5[-1,4]])

ind_mean = -np.log(np.mean(np.exp(-ind_all)))

delta_all = np.array([delta1[-1,4],delta2[-1,4],delta3[-1,4],delta4[-1,4],delta5[-1,4]])

delta_mean = -np.log(np.mean(np.exp(-delta_all)))

labels = ['WT','N501Y','K417N\nE484K\nN501Y','L452R\nE484Q', 'L452R\nT478K']
work_avg = [wt_mean,uk_mean,saf_mean,ind_mean,delta_mean]
colors = ['g','b','r','gray','orange']

plt.bar(labels,work_avg,capsize=20.0,color=colors)
plt.xticks(rotation=60)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.title('B38 + RBD',fontsize=16)
plt.ylabel('Nonequilibrium Work (kcal/mol)',fontsize=16)
plt.savefig('smd_plots/7BZ5_smd_work_avg.pdf',bbox_inches='tight')

```
