```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rbd = np.loadtxt('7BZ5_WT/production_new/rmsd/old/RBD_rmsd_500ns_RBD_aligned.dat')
antibody = np.loadtxt('7BZ5_WT/production_new/rmsd/old/Antibody_rmsd_500ns_RBD_aligned.dat')
antibody_bd = np.loadtxt('7BZ5_WT/production_new/rmsd/old/Antibody_binding_domain_rmsd_500ns_RBD_aligned.dat')
whole = np.loadtxt('7BZ5_WT/production_new/rmsd/old/whole_rmsd_500ns_RBD_aligned.dat')

t = np.arange(len(whole[:,0]))
t = t* 1E-4 #convert to us

plt.figure(figsize=(8,6))
#multiply by 10 to convert to Angstrom

plt.plot(t,antibody[:,1]*10.0,lw=0.5,color='g',label='Antibody (Full)')
plt.plot(t,antibody_bd[:,1]*10.0,lw=0.5,color='b',label='Antibody (Antigen binding domain)')
plt.plot(t,whole[:,1]*10.0,lw=0.5,color='orange',label='Whole complex')
plt.plot(t,rbd[:,1]*10.0,lw=0.5,color='r',label='RBD only')
#plt.legend(ncol=2,fontsize=16,markerscale=4.0)

plt.xlabel('Time ($\mu$s)',fontsize=20)
plt.ylabel('RMSD ($\AA$)',fontsize=20)
#plt.ylim(0,18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

plt.savefig('rmsd_figures/7bz5_wt.pdf',bbox_inches='tight')
```

```python
wt = np.loadtxt('7BZ5_WT/production_new/rmsd/RBD_rmsd_500ns_RBD_aligned.dat')
uk = np.loadtxt('7BZ5_N501Y/production_new/rmsd/RBD_rmsd_500ns_RBD_aligned.dat')
saf = np.loadtxt('7BZ5_South_African/production-new/rmsd/RBD_rmsd_500ns_RBD_aligned.dat')

t = np.arange(len(wt[:,0]))
t = t* 0.1 #convert to ns

t1 = np.arange(len(uk[:,0]))
t1 = t1* 0.1

plt.figure(figsize=(10,6))
#multiply by 10 to convert to Angstrom

plt.plot(t,wt[:,1]*10.0,lw=0.5,color='g',label='Wild type')
plt.plot(t1,uk[:,1]*10.0,lw=0.5,color='b',label='UK Variant')
plt.plot(t,saf[:,1]*10.0,lw=0.5,color='orange',label='SA Variant')

plt.legend(ncol=2,fontsize=16,markerscale=2.0)

plt.xlabel('Time (ns)',fontsize=20)
plt.ylabel('RMSD ($\AA$)',fontsize=20)
#plt.ylim(0,22)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.savefig('rmsd_figures/RBD_comparison.png',dpi=200,bbox_inches='tight')
```
