```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hb = []
with open("7BZ5_WT/production_new/H-bond-analysis/last_400ns_hbonds-details.dat") as f:
    i = 1
    for line in f:
        if i >= 3:
            y = line.split()
            labels = y[0]+'-'+y[1]
            prcnt = y[2]
            value = float(prcnt[0:-1])
            hb.append((labels,value))
        i += 1
hb.sort(key = lambda x: x[1], reverse=True)
labels = list(zip(*hb))[0]
values = list(zip(*hb))[1]
plt.bar(labels[0:15],values[0:15],color='r')
plt.xticks(rotation=90)
plt.ylabel('H-bond occupancy (%)',fontsize=16)
plt.yticks(fontsize=14)
plt.title('B38 + WT RBD',fontsize=16)
plt.ylim(0,60)
plt.savefig('hbond_figures/7BZ5_hbond_wt.pdf',bbox_inches='tight')
```

```python
l1 = np.loadtxt('7BZ5_WT/production_new/H-bond-analysis/last_400ns_hbonds.dat')
l2 = np.loadtxt('7BZ5_N501Y/production_new/H-bond-analysis/last_400ns_hbonds.dat')
l3 = np.loadtxt('7BZ5_South_African/production-new/H-bond-analysis/last_400ns_hbonds.dat')
l4 = np.loadtxt('7BZ5_Indian/production_new/H-bond-analysis/last_400ns_hbonds.dat')
l5 = np.loadtxt('7BZ5_delta/production/H-bond-analysis/last_400ns_hbonds.dat')

labels = ['WT','N501Y','K417N\nE484K\nN501Y','L452R\nE484Q','L452R\nT478K']
hbond_pt = [np.mean(l1[:,1]),np.mean(l2[:,1]),np.mean(l3[:,1]),np.mean(l4[:,1]),np.mean(l5[:,1])]
hbond_err = [np.std(l1[:,1]),np.std(l2[:,1]),np.std(l3[:,1]),np.std(l4[:,1]),np.std(l5[:,1])]
colors = ['g','b','r','gray','orange']

plt.bar(labels,hbond_pt,yerr=hbond_err,capsize=20.0,color=colors)
plt.xticks(rotation=60)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.title('B38 + RBD',fontsize=16)
plt.ylabel('Average Number of H-bonds',fontsize=16)
plt.savefig('hbond_figures/7BZ5_hbond_avg.pdf',bbox_inches='tight')
```
