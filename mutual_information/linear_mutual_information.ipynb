```r
library(bio3d)
library(igraph)
```

```r
#load files
dcdfile <- "heavy_100-200ns.dcd"
pdbfile <- "complex_heavy.pdb"

dcd <- read.dcd(dcdfile)
pdb <- read.pdb(pdbfile)

#get fitted trajectory 
ca.inds <- atom.select(pdb, elety="CA")

xyz <- fit.xyz(fixed=pdb$xyz, mobile=dcd,
               fixed.inds=ca.inds$xyz,
               mobile.inds=ca.inds$xyz)
trj <- xyz[,ca.inds$xyz]

#check dimensions
dim(xyz) == dim(dcd)
dim(trj)
dim(xyz)

#Calculate dynamical cross-correlation
cij<-dccm(xyz[,ca.inds$xyz],method="lmi",ncore=4)
plot(cij,zrange = c(0.0,1.0))


#Save cross correlation file
write.table(cij, file="mutual_info_100-200ns.txt", row.names=FALSE, col.names=FALSE)
```
